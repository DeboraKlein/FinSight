# Apresentação Executiva (Business Case)

O foco da apresentação é convencer a diretoria de que o projeto deve ser financiado e implementado, destacando o Retorno Ajustado ao Risco (RAROC) obtido.

## SLIDE 1: Título e Tese Central

### Modelo Preditivo de Intervenção Proativa

**Subtítulo**  
Maximizando o RAROC: Convertendo Risco de Rotativo em Ativo de Crédito Pessoal

---

**Tese Central**  
Intervir no **Estresse Agudo** do cliente (probabilidade ≥ **0.20**) com oferta de consolidação (crédito pessoal) para **mitigar 100% das perdas e garantir fidelização**.

## SLIDE 2: O Problema e a Solução

| Seção | O Problema Atual (Custo de Esperar) | Nossa Solução (Valor da Antecipação) |
| :--- | :--- | :--- |
| **Dor Financeira** | Clientes entrando no Rotativo (custo altíssimo) → Risco elevado de Perda (PDD) e Inadimplência. | Intervenção antes do default (**Threshold 0.20**), oferecendo uma solução de dívida mais barata (Empréstimo Pessoal). |
| **Modelo** | Abordagens reativas (Score Serasa/SPC). | **LightGBM otimizado** para **Recall (100%)** e **AUC de 0.7776**, focando em *features* de tendência de risco agudo. |

## SLIDE 3: Resultados Chave do Modelo (Proof of Concept)

O modelo foi validado em um período **Out-of-Time (OOT)** de estresse financeiro.

**1. Poder Preditivo (AUC):** **0.7776**  
* **Conclusão:** O modelo tem **alta capacidade de diferenciar** clientes de risco de clientes seguros (Superior ao Benchmark de **0.75**).

**2. Captura de Risco (Recall):** **100%**  
* **Conclusão:** Conseguimos **mitigar 100% do risco** - TODOS os clientes que entrariam no rotativo são capturados.

**3. Estratégia Otimizada (Threshold):** **0.20**  
* **Conclusão:** O *threshold* foi ajustado para maximizar a captura de risco, garantindo que o custo de Falsos Positivos (**Precision=36%**) seja convertido em **oportunidades de fidelização**.

## SLIDE 4: O Trade-off e o Custo Operacional (Maximizando o RAROC)

| Métrica | Risco Mitigado (Nosso Lucro) | Oportunidade Gerada (Fidelização) |
| :--- | :--- | :--- |
| **Verdadeiro Positivo (TP):** 15,273 clientes de risco | **Capital Econômico Reduzido:** Evita perdas por PDD e Inadimplência. | Conversão para Empréstimo Pessoal (Ativo de Maior Qualidade). |
| **Falso Positivo (FP):** 26,808 clientes seguros (**Precision de 36%**) | **Custo Operacional** (SAC, Gerentes). | **Investimento em Relacionamento:** A abordagem consultiva anula a insatisfação e aumenta a lealdade do cliente. |

**Mensagem Chave:** A **Alta Precisão do Modelo (AUC 0.7776)** justifica o investimento operacional, transformando um passivo (risco de rotativo) em um **ativo de portfólio sustentável**.

## SLIDE 5: Estrutura de Operacionalização (Fluxo de Valor)

Este slide mostra como o time de atendimento usa o modelo, garantindo escalabilidade via MLOps.

**Fluxo de Decisão:**
1. **Pipeline MLOps:** Modelo roda diariamente, gerando y_proba para toda a base.
2. **Decisão de Negócio:** Aplica a regra de y_proba ≥ **0.20**.
3. **Encaminhamento Otimizado:**
   * y_proba ≥ 0.50 (Risco Crítico) → **Gerente de Carteira/Telefone** (Alto Toque).
   * 0.20 ≤ y_proba < 0.50 (Risco Moderado) → **Canais Digitais** (Escala, Oferta Automatizada).

**Top 5 Preditores do Modelo:**
1. **limite_cartao** (354) - Exposição total ao crédito
2. **scr_tcr_implicita** (333) - Risco sistêmico
3. **utilizacao_limite_media_3m** (297) - Padrão de uso
4. **scr_crescim_divida_3m** (208) - Aceleração da dívida
5. **gasto_crescim_3m** (186) - Pressão de consumo

**Mensagem Chave**  
"O alerta não é sobre cobrança; é sobre a **proposta de valor da FinTech** para a saúde financeira do cliente."
