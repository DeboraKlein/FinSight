## Apresentação Executiva (Business Case)

O foco da apresentação é convencer a diretoria de que o projeto deve ser financiado e implementado, destacando o Retorno Ajustado ao Risco (RAROC) obtido.



# SLIDE 1: Título e Tese Central

## Modelo Preditivo de Intervenção Proativa

### Subtítulo
Maximizando o $\text{RAROC}$: Convertendo Risco de Rotativo em Ativo de Crédito Pessoal

---

### Tese Central
Intervir no **Estresse Agudo** do cliente (probabilidade $\ge \mathbf{0.20}$) com oferta de consolidação (crédito pessoal) para **mitigar perdas e garantir fidelização**.

# SLIDE 2: O Problema e a Solução

| Seção | O Problema Atual (Custo de Esperar) | Nossa Solução (Valor da Antecipação) |
| :--- | :--- | :--- |
| **Dor Financeira** | Clientes entrando no Rotativo (custo altíssimo) $\rightarrow$ Risco elevado de Perda ($\text{PDD}$) e Inadimplência. | Intervenção antes do default ($\text{Threshold} 0.20$), oferecendo uma solução de dívida mais barata (Empréstimo Pessoal). |
| **Modelo** | Abordagens reativas (Score Serasa/SPC). | **LightGBM otimizado** para $\text{Recall}$ ($\mathbf{89\%}$) e **$\text{AUC}$ de $\mathbf{0.7732}$**, focando em *features* de tendência de risco agudo. |

# SLIDE 3: Resultados Chave do Modelo (Proof of Concept)

O modelo foi validado em um período **Out-of-Time (OOT)** de estresse financeiro.

1. **Poder Preditivo ($\text{AUC}$):** $\mathbf{0.7732}$
   * **Conclusão:** O modelo tem **alta capacidade de diferenciar** clientes de risco de clientes seguros (Superior ao Benchmark de $\mathbf{0.75}$).

2. **Captura de Risco ($\text{Recall}$):** $\mathbf{89\%}$
   * **Conclusão:** Conseguimos **mitigar o risco** em quase **9 de cada 10 clientes** antes que o rotativo se consolide.

3. **Estratégia Otimizada ($\text{Threshold}$):** $\mathbf{0.20}$
   * **Conclusão:** O *threshold* foi ajustado para maximizar a captura de risco, garantindo que o custo de Falsos Positivos ($\text{Precision}=32\%$) seja convertido em **oportunidades de fidelização**.

# SLIDE 4: O Trade-off e o Custo Operacional (Maximizando o $\text{RAROC}$)

| Métrica | Risco Mitigado (Nosso Lucro) | Oportunidade Gerada (Fidelização) |
| :--- | :--- | :--- |
| **Verdadeiro Positivo (TP):** Clientes de Risco que o Modelo ACERTA | **$\text{Capital Econômico}$ Reduzido:** Evita perdas por $\text{PDD}$ e $\text{Inadimplência}$. | Conversão para Empréstimo Pessoal (Ativo de Maior Qualidade). |
| **Falso Positivo (FP):** Clientes Seguros que o Modelo CLASSIFICA como Risco ($\text{Precision}$ de $32\%$) | **Custo Operacional** (SAC, Gerentes). | **Investimento em Relacionamento:** A abordagem consultiva anula a insatisfação e aumenta a lealdade do cliente. |

**Mensagem Chave:** A **Alta Precisão do Modelo ($\text{AUC}$)** justifica o investimento operacional, transformando um passivo (risco de rotativo) em um **ativo de portfólio sustentável**.

# SLIDE 5: Estrutura de Operacionalização (Fluxo de Valor)

Este slide mostra como o time de atendimento usa o modelo, garantindo escalabilidade via $\text{MLOps}$.

### Fluxo de Decisão:
1. **Pipeline $\text{MLOps}$:** Modelo roda diariamente, gerando $y\_proba$ para toda a base.
2. **Decisão de Negócio:** Aplica a regra de $y\_proba \ge \mathbf{0.20}$.
3. **Encaminhamento Otimizado:**
    * $y\_proba \ge 0.50$ (Risco Crítico) $\rightarrow$ **Gerente de Carteira/Telefone** (Alto Toque).
    * $0.20 \le y\_proba < 0.50$ (Risco Moderado) $\rightarrow$ **Canais Digitais** (Escala, Oferta Automatizada).

### Mensagem Chave
"O alerta não é sobre cobrança; é sobre a **proposta de valor da FinTech** para a saúde financeira do cliente."
