## Sustentabilidade Financeira: RAROC (Risk-Adjusted Return on Capital)

#### Do Score Estatístico ao Retorno Ajustado ao Risco

Esta seção calcula o impacto financeiro da intervenção preditiva e define o **threshold ideal de atuação** para a FinTech, garantindo que a mitigação de risco maximize o retorno do capital investido.

---

### 1. Fundamentação: Por que o RAROC?

**O RAROC (Retorno Ajustado ao Risco sobre o Capital) é a métrica essencial para finanças.** Em modelos de crédito, não basta ter um modelo preciso ($AUC$ alto); é preciso que a estratégia de intervenção seja **lucrativa**.

O RAROC compara o retorno esperado de uma operação com o risco que essa operação consome do capital da empresa. O objetivo do Machine Learning é **maximizar o RAROC**, encontrando o ponto ótimo onde o retorno é alto, e a perda (risco) é minimizada pela intervenção preditiva.

**Fórmula Simplificada:**

$$\\text{RAROC} = \\frac{\\text{Retorno Ajustado ao Risco}}{\\text{Capital Econômico (Risco)}}$$

---

### 2. Componentes da Simulação do RAROC

Para calcular o RAROC, a simulação se baseia em quatro componentes principais:

#### 2.1. Capital Econômico (CE):

Representa o capital necessário para absorver as **perdas não esperadas (Unexpected Loss - UL)**. É o capital que deve ser reservado para que a FinTech sobreviva a eventos de estresse (como a crise simulada). O CE é a variável que o ML tenta **reduzir** ao evitar o default.

#### 2.2. Perda Esperada (PE - Expected Loss):

A perda que a FinTech espera absorver em um ano normal (ex: perdas operacionais e defaults esperados). O ML deve **manter a PE controlada**.

#### 2.3. Receita (Revenue):

O valor das receitas geradas pelos clientes que **não foram para o rotativo** após a intervenção (empréstimo pessoal). O ML tenta **aumentar a Receita** ao reter bons clientes e converter dívida cara em dívida rentável.

#### 2.4. Retorno Ajustado ao Risco:

$$\\text{Retorno Ajustado} = \\text{Receita} - \\text{Custo Operacional} - \\text{Perda Esperada (PE)}$$

---

### 3. A Simulação: Efeito do Threshold na Carteira

A simulação do RAROC avalia o impacto financeiro da intervenção preditiva para cada possível *threshold* (limiar de probabilidade de risco) de **0.01 a 0.50**. 

| Threshold | Ação | Efeito no ML (Recall vs Precision) |
| :---: | :--- | :--- |
| **Mais Baixo (ex: 0.10)** | Intervenção Agressiva (muitos alertas). | **Maximiza o Recall** (mitiga quase todo o risco), mas **Minimiza o Precision** (alto custo operacional com Falsos Positivos). |
| **Mais Alto (ex: 0.40)** | Intervenção Conservadora (poucos alertas). | **Maximiza o Precision** (baixo custo operacional), mas **Minimiza o Recall** (alto risco de perdas não mitigadas). |

O cálculo do RAROC busca o ponto de inflexão onde o **Retorno Ajustado** é otimizado em relação ao **Capital Econômico**.

---

### 4. O Objetivo do ML: Conectando Precisão e Retorno

O modelo de Machine Learning (LightGBM) é a ferramenta que determina a probabilidade de risco do cliente. A **precisão do modelo** define o limite máximo de RAROC que a FinTech pode alcançar.

| Métrica de ML | Valor Validado (OOT) | Significado para o RAROC |
| :--- | :--- | :--- |
| **Poder Preditivo (AUC)** | **0.7281** | Comprova a capacidade do modelo de discriminar o cliente bom do ruim, definindo o teto de otimização de risco. |
| **Threshold Otimizado** | **0.20** | O limiar que resultou no maior RAROC (máximo retorno ajustado ao risco) na simulação. | 

**A Importância do 0.20:** Ao escolher este *threshold*, a FinTech está aceitando um *trade-off* onde a mitigação do risco (alto Recall) é prioridade, pois o custo da perda financeira é superior ao custo de intervir em Falsos Positivos. 

---

### 5. Conclusão Estratégica: O Retorno Garantido pela Intervenção

O projeto é finalizado com a validação do **resultado de negócio**, que é a prova da sustentabilidade financeira da solução preditiva.

#### 5.1. Performance Validada (Threshold 0.20):

| Métrica | Valor Final | Significado para a Sustentabilidade |
| :---: | :---: | :--- |
| **Threshold de Ação** | **0.20** | Ponto de corte que maximiza o retorno ajustado ao risco (RAROC). |
| **Recall (Clientes Salvos)** | **89%** | **Alto Retorno:** O modelo garante que **9 em cada 10 clientes** em risco de default sejam interceptados antes da perda, maximizando o Retorno Ajustado. |
| **Precision (Custo Operacional)** | **32%** | **Risco Mitigado:** A alta taxa de Falsos Positivos (68%) é o custo necessário para atingir o Recall de 89%, e é compensada pela estratégia de relacionamento (oferta de crédito mais barato). |

#### 5.2. O Impacto Financeiro da Estratégia:

Ao usar o modelo LightGBM com o *threshold* de **0.20**, a FinTech muda a narrativa do risco:

* **Redução do Capital Econômico (CE):** A alta capacidade preditiva (AUC 0.7281) permite que a FinTech reserve menos capital para perdas inesperadas, liberando capital para novas operações.
* **Maximização do Retorno:** A intervenção proativa evita que os clientes se tornem PDD (Provisão para Devedores Duvidosos) caros, convertendo-os em ativos de empréstimo pessoal, aumentando a receita sustentável.

O projeto comprova que a inteligência preditiva é a chave para a sustentabilidade e crescimento lucrativo no mercado de crédito.
